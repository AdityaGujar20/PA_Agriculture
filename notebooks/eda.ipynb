```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

def clean_data(file_path):
    df = pd.read_csv(file_path)

    # change thhe data type of date column to datetime
    df['date'] = pd.to_datetime(df['date'])

    df['month'] = df['date'].dt.month()
    df['year'] = df['date'].dt.year()

    # fill missing values
    values = {
        "soil_N": df['soil_N'].mean(),
        "rainfall_mm": df['rainfall_mm'].mean(),
        "fertilizer_kg_per_ha": df['fertilizer_kg_per_ha'].mean()
    }
    df.fillna(value = values)

    # drop month and day_of_year
    df.drop(columns=['month', 'day_of_year'], inplace = True)

    return df.head()




```

```python
clean_data(file_path="../data/Synthetic_Farming_Dataset_With_Seasonality_And_Challenge - Synthetic_Farming_Dataset_With_Seasonality_And_Challenge.csv")
```
